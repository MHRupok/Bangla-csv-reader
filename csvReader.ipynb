```python
import numpy as np
import pandas as pd
```

```python
def load_dataset(filename):
    df = pd.read_csv(filename,encoding = "utf-8",names = ["Sentence", "Intent"])
    print(df.head())
    intent = df["Intent"]
    unique_intent = list(set(intent))
    sentences = list(df["Sentence"])
  
    return (intent, unique_intent, sentences)
```
